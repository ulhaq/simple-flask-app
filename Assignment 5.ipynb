Code that exposes an endpoint at the server root 

```python
from flask import Flask, Response, request

app = Flask(__name__)

@app.route('/', methods=['POST'])
def index():
    SERVER_PASSWORD = '11749'

    data = request.get_json()

    if data['client_password'] == SERVER_PASSWORD:
        return Response('{"success": "true"}', status=200, mimetype='application/json')
    else:
        return Response('{"success": "false"}', status=403, mimetype='application/json')
```

```python
import string
import requests
import json
from itertools import permutations

def pass_check(password):
    url = "http://127.0.0.1:5000/"
    headers = {"Content-type" : "application/json"}
    payload = json.dumps({"client_password": password})

    r = requests.post(url, headers=headers, data=payload)

    return r, r.text

data = permutations(string.printable, 2)

for i in data:
    x = str(ord(i[0]))+str(ord(i[1]))
    
    print(i,x,pass_check(x))
```
